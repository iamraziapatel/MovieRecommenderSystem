```python
import ast  # module (Abstract Syntax Trees) is a built-in module that provides facilities to work with the abstract syntax tree of Python source code


# helper function to remove only specific  "name": from the genre dictionary
def convert(obj): 
    L=[]
    for i in ast.literal_eval(obj):  # converts string type to list type ,The literal_eval() function safely evaluates an expression node or a string containing a Python literal or container display.
        name = i['name'].replace(" ", "")
        L.append(name)    
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
movies['keywords']=movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast'][0]
```

```python
def convert3(obj): 
    L=[]
    counter = 1    
    for i in ast.literal_eval(obj):  # converts string type to list type ,The literal_eval() function safely evaluates an expression node or a string containing a Python literal or container display.
        if counter != 3:  
            name = i['name'].replace(" ", "")
            L.append(name)    
    return L
```

```python
movies['cast']=movies['cast'].apply(convert)
```

```python
def fetchDirector(obj):
    L=[]
    for i  in ast.literal_eval(obj):
        if (i["job"] == "Director"):
            name=i["name"].replace(" ","")
            L.append(name)
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetchDirector)
```

```python
movies['overview'] = movies['overview'].fillna(movies['title'])

movies.head()
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies["tags"]=movies["overview"]+movies["genres"]+movies["keywords"]+movies["crew"]
newdf_movies=movies[['movie_id','title','tags']]
newdf_movies.head()
newdf_movies['tags'][0]
```

```python

newdf_movies['tags']=newdf_movies['tags'].apply(lambda x:" ".join(x))


```

```python
newdf_movies['tags']=newdf_movies['tags'].apply(lambda x:x.lower())
```

```python
from nltk.stem import PorterStemmer

# Custom tokenizer function with stemming
def custom_tokenizer(text):
    stemmer = PorterStemmer()
    words = text.split()  # Split the text into words
    stemmed_words = [stemmer.stem(word) for word in words if not word.isdigit()]
    return " ".join(stemmed_words)  # Return the stemmed words as a single string

# Convert the "tags" column to lowercase and apply stemming
newdf_movies['tags'] = newdf_movies['tags'].apply(lambda x: custom_tokenizer(x.lower()))
```

```python
import re
from sklearn.feature_extraction.text import CountVectorizer

# # Custom tokenizer function that filters out numerical values
# def custom_tokenizer(text):
#     # Use regular expression to split the text into words
#     words = re.findall(r'\b\w+\b', text.lower())
#     # Filter out numerical values from the words list
#     words = [word for word in words if not word.isdigit()]
#     return words

# Create the CountVectorizer with the custom tokenizer
# cv = CountVectorizer(max_features=5000, stop_words='english', tokenizer=custom_tokenizer)
cv = CountVectorizer(max_features=5000, stop_words='english')
# Fit and transform the 'tags' column using the custom tokenizer
vectors = cv.fit_transform(newdf_movies["tags"]).toarray()

# Print the first row of vectors (numerical representation of the tags for the first movie)
print(vectors[0])

# Get the feature names after fitting the CountVectorizer
print(cv.get_feature_names_out())
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vectors).shape #Cosine similarity gives us a matrix since it calculates the distance with each vector hence the diagonal will always be 1
```

```python
similarity =cosine_similarity(vectors)

sorted(similarity[0],reverse=True) #so we are loosing the index 
```

The enumerate() function in Python is used to iterate over a sequence (e.g., a list, tuple, or string) while keeping track of the index of the current item. It returns a sequence of tuples, where each tuple contains the index and the corresponding element from the original sequence.

When you use list(enumerate(similarity)), it means you are applying the enumerate() function to the similarity object, which is typically a list or an iterable. The enumerate() function will then iterate over the elements in the similarity list and return a list of tuples, where each tuple contains the index of the element and the element itself.:

```python

def Recommend(movie):
    movie_index=newdf_movies[newdf_movies['title'] == movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(newdf_movies.iloc[i[0]].title)
    
newdf_movies.head()
```

```python
Recommend('The Lone Ranger')
```
